```python
import os
import pandas as pd
from datetime import date
```

```python
today = date.today()
end_date = today
```

# SCRAPPING USING SEARCH TERM

```python
search_term = 'bully'
from_date = '2019-01-01'
```

```python
max_results = 100
```

```python
extracted_tweets = "snscrape --format '{content!r}'"+ f" --max-results {max_results} --since {from_date} twitter-search '{search_term} until:{end_date}' > search_term.csv"
os.system(extracted_tweets)
if os.stat("search_term.csv").st_size == 0:
  print('No Tweets found')
else:
  df = pd.read_csv('search_term.csv', names=['content'])
```

# SCRAPPING USING PUBLIC ACCOUNT

```python
user_name = "BarackObama"
user_tweets = "snscrape --format '{content!r}'"+ f" --max-results {max_results} --since {from_date} twitter-user '{user_name} until:{end_date}' > user-tweets.csv"
os.system(user_tweets)
if os.stat("user-tweets.csv").st_size == 0:
  print('No Tweets found')
else:
  df = pd.read_csv('user-tweets.csv', names=['content'])
```
